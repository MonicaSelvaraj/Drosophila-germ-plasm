# Density Based Clustering and Validation of 3D data

```python
#Importing libraries
from sklearn.datasets.samples_generator import make_blobs

import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
%matplotlib notebook

import numpy as np
import hdbscan
```

```python
#Generating sample 3D data 
data, labels = make_blobs(n_samples=100, centers=3, n_features=3)
x,y,z = np.hsplit(data,3)
x = np.concatenate(x,axis=0);y = np.concatenate(y,axis=0);z = np.concatenate(z,axis=0)

#Fig1 - Visualizing sample 3D data
fig = plt.figure( )
plt.style.use('dark_background')
ax = fig.add_subplot(1,1,1, projection = '3d')
ax.grid(False)
ax.scatter (x, y, z, c = 'g', marker='o', s=1, linewidths=2)
ax.set_title('Sample data for clustering')
ax.set_xlabel ('x, axis')
ax.set_ylabel ('y axis')
ax.set_zlabel ('z axis')
plt.show()
```

```python
plt.close()
#Training HDBSCAN model
model = hdbscan.HDBSCAN(min_cluster_size = 20, gen_min_span_tree = True)
model.fit(data)
print(model)
```

```python
plt.close()
model.condensed_tree_.plot()
```

```python
plt.close()
model.condensed_tree_.plot(select_clusters=True)
```

```python
plt.close()
model.minimum_spanning_tree_.plot(edge_cmap = 'viridis', edge_alpha =0.6, node_size = 80, edge_linewidth =2)
```
